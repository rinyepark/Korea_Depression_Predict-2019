* DI1_pr : 9(892)

#### 1: 751 -> 744

```python
data_del3 = data_del2.copy()
```

```python
data_3_col = list(data_del3.columns)
col3_d9 = []
col3_d999 = []
for i in range(len(data_3_col)):
    tmp = data_3_col[i][-2:]
    temp = data_3_col[i]
    if 'dg' in tmp: 
        col3_d9.append(temp)
    elif 'pr' in tmp or 'pt' in tmp:
        col3_d9.append(temp)
    elif 'ag' in tmp:
        col3_d999.append(temp)
```

```python
for i in range(len(col3_d9)):
    data_del3 = data_del3[data_del3[col3_d9[i]] != 9]

for i in range(len(col3_d999)):
    data_del3 = data_del3[data_del3[col3_d999[i]] != 999]
```

```python
print(sorted(Counter(data_del3['Is_Depressed']).items()))
```
